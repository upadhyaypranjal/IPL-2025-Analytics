```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Clares10@",  # put your password
    database="ipl2025"
)

pp_query = """
SELECT striker AS player,
       SUM(runs_of_bat) AS runs_powerplay
FROM deliveries_2025
WHERE `over` BETWEEN 1 AND 6
GROUP BY striker
ORDER BY runs_powerplay DESC
LIMIT 5;
"""

pp_df = pd.read_sql(pp_query, conn)
print(pp_df)

plt.figure(figsize=(8,4))
plt.bar(pp_df["player"], pp_df["runs_powerplay"])
plt.title("Top 5 Powerplay Scorers - IPL 2025")
plt.xlabel("Players")
plt.ylabel("Runs")

plt.savefig("../data_processed/visuals/powerplay_top5.png", dpi=300, bbox_inches="tight")


plt.show()
conn.close()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Reconnect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Clares10@", 
    database="ipl2025"
)

bowler_query = """
SELECT bowler,
       COUNT(*) AS wickets
FROM deliveries_2025
WHERE wicket_type IS NOT NULL
GROUP BY bowler
ORDER BY wickets DESC
LIMIT 5;
"""

bowler_df = pd.read_sql(bowler_query, conn)
print(bowler_df)

plt.figure(figsize=(8,4))
plt.bar(bowler_df["bowler"], bowler_df["wickets"])
plt.title("Top 5 Wicket Takers - IPL 2025")
plt.xlabel("Bowler")
plt.ylabel("Wickets Taken")
plt.show()

conn.close()

```
